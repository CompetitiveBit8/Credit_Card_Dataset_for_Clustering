```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
import warnings
```

```python
warnings.filterwarnings(action='ignore')
df = pd.read_csv('CC_GENERAL.csv')
df
```

### Data Engineering - (Adjusting for Null values)

```python
X = df.drop(columns=df[['CUST_ID', 'TENURE']])
X.isnull().sum()
```

```python
X.fillna(method='ffill', inplace=True)
X.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

### Scaling

```python
scaler = MinMaxScaler()
X_transformed = scaler.fit_transform(X)
```

### Initial Clustering

```python
km = KMeans(n_clusters=4)
y_predict = km.fit_predict(X)
pd.DataFrame(y_predict).value_counts()
```

### Checking through a range of K values to find the Best

```python
K_range = range(1, 10)
sse = []

for k in K_range:
    km = KMeans(n_clusters=k)
    km.fit_predict(X)
    sse.append(km.inertia_)
```

### Plotting of K values against the Sum of Squares

```python
plt.figure(figsize=(8, 6))
plt.plot(K_range, sse)
plt.xlabel('sse')
plt.ylabel('K_range')
plt.title('Sum of Square')
plt.grid()
plt.show()
```

###  Predicting using the K value derived form the graphical analysis

```python
km = KMeans(n_clusters=2)
y_predict = km.fit_predict(X)
y_predict
```

### Table presentation with Label added

```python
df['Labels'] = y_predict
df
```
